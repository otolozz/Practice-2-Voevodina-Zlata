### <font color = "#48D1CC"> *Context*

Музыкальная терапия, или МТ, - это использование музыки для борьбы со стрессом, улучшения настроения и общего психического здоровья человека. МТ также признана научно обоснованной практикой, использующей музыку в качестве катализатора выработки "гормонов счастья", таких как окситоцин.

Набор данных направлен на то, чтобы проверить гипотезы и определить, какие существуют корреляции, если таковые имеются, между музыкальным выбором человека и его состоянием психического здоровья. 

### <font color = "#48D1CC"> *Content*

#### Блок 1: Музыкальные жанры

Респонденты отвечают на общие вопросы, касающиеся музыкального фона и привычек прослушивания.:

* Никогда
* Редко
* Иногда
* Очень часто

#### Блок 2: Психическое здоровье

Респонденты оценивают тревогу, депрессию, бессонницу и ОКР по шкале от 0 до 10, где:

* 0 - Я не испытываю этого
* 10 - Я испытываю это регулярно, постоянно / или чересчур часто.

### <font color = "#48D1CC"> *Variables*

* Frequency [Country] - How frequently the respondent listens to country music
* Frequency [EDM] - How frequently the respondent listens to EDM music
* Frequency [Folk] - How frequently the respondent listens to folks music
* Frequency [Gospel] - How frequently the respondent listens to Gospel music
* Frequency [Hip hop] - How frequently the respondent listens to hip hop music
* Frequency [Jazz] - How frequently the respondent listens to jazz music
* Frequency [K pop] - How frequently the respondent listens to K pop music
* Frequency [Latin] - How frequently the respondent listens to Latin music
* Frequency [Lofi] - How frequently the respondent listens to lofi music
* Frequency [Metal] - How frequently the respondent listens to metal music 
* Frequency [Pop] - How frequently the respondent listens to pop music
* Frequency [R&B] - How frequently the respondent listens to R&B music
* Frequency [Rap] - How frequently the respondent listens to rap music
* Frequency [Rock] - How frequently the respondent listens to rock music
* Frequency [Video game music] - How frequently the respondent listens to video game music
* Anxiety - Self-reported anxiety, on a scale of 0-10
* Depression - Self-reported depression, on a scale of 0-10
* Insomnia - Self-reported insomnia, on a scale of 0-10
* OCD - Self-reported OCD, on a scale of 0-10
* Music effects - Does music improve/worsen respondent's mental health conditions?
* Permissions - Permissions to publicize data

Пусть для уровней признаков ментального здоровья будет:
* 0 - отличное ментальное здоровье
* 1-3 - нормальное ментальное здоровье
* 4-7 - плохое ментальное здоровье
* 8-10 - очень плохое ментальное здоровье

```python
#создадим условие для уровня ментального здоровья

def get_mh_cat(mh):
    if mh <= 0 :
        return 'Отличное ментальное здоровье'
    elif mh <= 3:
        return 'Нормальное ментальное здоровье'
    elif mh <= 7:
        return 'Плохое ментальное здоровье'
    elif mh <= 10:
        return 'Очень плохое ментальное здоровье'
    else: 
        return mh

mumehel['Anxiety'].apply(get_mh_cat)
mumehel['Depression'].apply(get_mh_cat)
mumehel['Insomnia'].apply(get_mh_cat)
mumehel['OCD'].apply(get_mh_cat)
```

```python
#добавим новую категориальную переменную в таблицу
mumehel['Anx, cat'] = mumehel['Anxiety'].apply(get_mh_cat)
mumehel['Dep, cat'] = mumehel['Depression'].apply(get_mh_cat)
mumehel['Ins, cat'] = mumehel['Insomnia'].apply(get_mh_cat)
mumehel['OCD, cat'] = mumehel['OCD'].apply(get_mh_cat)

#переведем новую переменную в категориальную
mumehel['Anx, cat'] = pd.Categorical(mumehel['Anx, cat'])
mumehel['Dep, cat'] = pd.Categorical(mumehel['Dep, cat'])
mumehel['Ins, cat'] = pd.Categorical(mumehel['Ins, cat'])
mumehel['OCD, cat'] = pd.Categorical(mumehel['OCD, cat'])

mumehel.head()
```

```python
#посмотрим распределение ответов для признаков ментального состояния

fig, ax = plt.subplots(2, 2, figsize=(20, 10))

ax[0][0].bar(mumehel['Anx, cat'].value_counts().index, mumehel['Anx, cat'].value_counts(), color='#48D1CC')
ax[0][1].bar(mumehel['Dep, cat'].value_counts().index, mumehel['Dep, cat'].value_counts(), color='#48D1CC')
ax[1][0].bar(mumehel['Ins, cat'].value_counts().index, mumehel['Ins, cat'].value_counts(), color='#48D1CC')
ax[1][1].bar(mumehel['OCD, cat'].value_counts().index, mumehel['OCD, cat'].value_counts(), color='#48D1CC')

ax[0][0].set_xlabel ('Уровень тревожности', fontsize = 15)
ax[0][1].set_xlabel ('Уровень депрессии', fontsize = 15)
ax[1][0].set_xlabel ('Уровень бессоннницы', fontsize = 15)
ax[1][1].set_xlabel ('Уровень ОКР', fontsize = 15)

ax[0][0].set_xticklabels(mumehel['Anx, cat'].value_counts().index, rotation=8, ha='right')
ax[0][1].set_xticklabels(mumehel['Dep, cat'].value_counts().index, rotation=8, ha='right')
ax[1][0].set_xticklabels(mumehel['Ins, cat'].value_counts().index, rotation=8, ha='right')
ax[1][1].set_xticklabels(mumehel['OCD, cat'].value_counts().index, rotation=8, ha='right')

for axis in ax.ravel ():
    axis.set_ylabel ('Количество респондентов', fontsize = 15)


fig.suptitle('Ментальное состояние респондентов', fontsize = 20)
```

```python
#создадим и применем условие частот прослушивания жанров

def frequency (fr):
    if fr == 'Never' :
        return 0
    elif fr == 'Rarely':
        return 1
    elif fr == 'Sometimes':
        return 2
    elif fr == 'Very frequently':
        return 3
    else: 
        return fr

mumehel['Frequency [Classical]'] = mumehel['Frequency [Classical]'].apply(frequency)
mumehel['Frequency [Country]'] = mumehel['Frequency [Country]'].apply(frequency)
mumehel['Frequency [EDM]'] = mumehel['Frequency [EDM]'].apply(frequency)
mumehel['Frequency [Folk]'] = mumehel['Frequency [Folk]'].apply(frequency)
mumehel['Frequency [Gospel]'] = mumehel['Frequency [Classical]'].apply(frequency)
mumehel['Frequency [Hip hop]'] = mumehel['Frequency [Country]'].apply(frequency)
mumehel['Frequency [Jazz]'] = mumehel['Frequency [EDM]'].apply(frequency)
mumehel['Frequency [K pop]'] = mumehel['Frequency [Folk]'].apply(frequency)
mumehel['Frequency [Latin]'] = mumehel['Frequency [Classical]'].apply(frequency)
mumehel['Frequency [Lofi]'] = mumehel['Frequency [Country]'].apply(frequency)
mumehel['Frequency [Metal]'] = mumehel['Frequency [EDM]'].apply(frequency)
mumehel['Frequency [Pop]'] = mumehel['Frequency [Folk]'].apply(frequency)
mumehel['Frequency [R&B]'] = mumehel['Frequency [Classical]'].apply(frequency)
mumehel['Frequency [Rap]'] = mumehel['Frequency [Country]'].apply(frequency)
mumehel['Frequency [Rock]'] = mumehel['Frequency [EDM]'].apply(frequency)
mumehel['Frequency [Video game music]'] = mumehel['Frequency [Folk]'].apply(frequency)
```

```python
#регрессия на проверку тревожности

X = mumehel['Hours per day']
y = mumehel['Anxiety']

res = scipy.stats.linregress(X, y)
print(res)
```

```python
r_sq = res.rvalue ** 2
print(r_sq)
```

Мы видим, что r2 слишком маленький, что говорит о том, что модель может объяснить лишь 0.2% разброса данных. 

```python
#регрессия на проверку депрессии

X = mumehel['Hours per day']
y = mumehel['Depression']

res1 = scipy.stats.linregress(X, y)
print(res1)
```

```python
r_sq1 = res1.rvalue ** 2
print(r_sq1)
```

В случае с показателем депрессии, r2 выше, что означает, что данная модель лучше объясняет разброс данных.
